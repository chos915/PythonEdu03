## 사칙연산

```python
score1 = [10, 15, 24, 169, 90]
score2 = [30, 40, 70, 50, 100]

score1 + score2
```

```python
score1_arr = np.array(score1)
score2_arr = np.array(score2)
                    
result = score1_arr + score2_arr
print(np.min(result)) # 40, 55, 94, 219, 190
```

```python
print(np.max(result)) # 40, 55, 94, 219, 190
print(np.sum(result)) # 40, 55, 94, 219, 190
```

## 배열 연산
- 덧셈 : np.add()
- 뺄셈 : np.subtract()
- 곱셈 : np.multiply()
- 나눗셈 : np.divide()

```python
import numpy as np

array_01 = np.array([1, 2, 3])
array_02 = np.array([10, 20, 30])

# 덧셈
result = np.add(array_01, array_02)
result
```

```python
# 뺄셈
result = np.subtract(array_01, array_02)
result
```

```python
# 곱셈
result = np.multiply(array_01, array_02)
result
```

```python
# 나눗셈
result = np.divide(array_01, array_02)
result
```

```python
array_01 = np.array([1, 2, 3])
array_02 = np.array([1, 2, 3])

result = np.power(array_01, array_02)
print(result)
```

## 소숫점 정렬
- 소수점 절삭삭

- 코드는 실행속도를 체크해야 한다.

```python
%%time

temp_arr = np.trunc([-1.23, 1.23])
print(temp_arr)
```

```python
%%time

temp_arr = np.fix([-1.23, 1.23])
print(temp_arr)
```

- 소숫점 자리에서 반올림

```python
temp_arr = np.around([-1.123131, 1.123131], 3)
temp_arr
```

```python
temp_arr = np.round([-1.123131, 1.123131], 3)
temp_arr
```

- 소숫점 모두

```python
import time

# Start time
start_time = time.time()

temp_arr = np.floor([-1.23456, 1.23456])
print(temp_arr)

# End time
end_time = time.time()

# Execution time
execution_time = end_time - start_time

print("Execution time:", execution_time)
```

```python
import time

# Start time
start_time = time.time()

temp_arr = np.ceil([-1.23456, 1.23456])
print(temp_arr)

# End time
end_time = time.time()

# Execution time
execution_time = end_time - start_time

print("Execution time:", execution_time)
```

## 조건문
- where : 조건식 1개일 때 활용
- select : 다중조건일 때 활용

```python
temp_arr = np.arange(10)
temp_arr
```

- np.where

```python
np.where(temp_arr < 5, temp_arr, 10 * temp_arr)
```

- np.select

```python
temp_arr = np.arange(10)
cond_list = [temp_arr > 5, temp_arr < 2]
choice_list = [temp_arr * 3, temp_arr + 10]

np.select(cond_list, choice_list, default=temp_arr)
```
